```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score

#membuat data harga tanah
np.random.seed(42)

# Luas tanah (m^2)
X = np.random.randint(50, 600, 80).reshape(-1, 1)

# Rumus harga tanah realistis
# Harga = 500k * luas + 0.8 * luas^2 + sedikit noise
y = 500_000 * X.flatten() + 0.8 * (X.flatten()**2) + np.random.randint(-50_000_000, 50_000_000, 80)

#model polinomial
degree = 2
poly = PolynomialFeatures(degree=degree)
X_poly = poly.fit_transform(X)

model = LinearRegression()
model.fit(X_poly, y)

# Prediksi untuk plot
X_test = np.linspace(50, 600, 200).reshape(-1, 1)
X_test_poly = poly.transform(X_test)
y_pred = model.predict(X_test_poly)

#mengevaluasi model
mse = mean_squared_error(y, model.predict(X_poly))
rmse = np.sqrt(mse)
r2 = r2_score(y, model.predict(X_poly))

print("Degree =", degree)
print("MSE  =", mse)
print("RMSE =", rmse)
print("R^2  =", r2)

#membuat grafik
plt.figure(figsize=(10, 6))
plt.scatter(X, y, label="Data Observasi")
plt.plot(X_test, y_pred, label=f"Polynomial Degree {degree}", linewidth=3)
plt.title("Polynomial Regression - Harga Tanah")
plt.xlabel("Luas Tanah (m^2)")
plt.ylabel("Harga (Rupiah)")
plt.legend()
plt.grid(True)
plt.show()

#memprediksi
luas_sample = np.array([[50], [100], [200], [300], [500]])
luas_poly = poly.transform(luas_sample)
predicted = model.predict(luas_poly)

for luas, harga in zip(luas_sample.flatten(), predicted):
    print(f"Luas {luas} m2  Prediksi harga  Rp {harga:,.0f}")

```
